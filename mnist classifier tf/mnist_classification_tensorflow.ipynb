```python
import tensorflow as tf
import numpy as np
import scipy
from sklearn import model_selection
np.set_printoptions(precision=3, suppress=True)
print("Num GPUs Available: ", len(tf.config.list_physical_devices('GPU')))
print(tf.__version__)
```

```python
from tensorflow.keras.layers import Dense
from tensorflow.keras import Sequential

model = Sequential()
model.add(Dense(28*28*2, activation='relu'))
model.add(Dense(10, activation='softmax'))

model.compile(loss=tf.keras.metrics.categorical_crossentropy,
              optimizer=tf.optimizers.SGD(1),
              metrics=['accuracy'])
```

```python
(x_train, y_train), (x_test, y_test) = tf.keras.datasets.mnist.load_data()

x_train=tf.keras.utils.normalize(x_train,axis=1)
x_test=tf.keras.utils.normalize(x_test,axis=1)

x_train = x_train.reshape((60000, 1, 28*28))
x_test = x_test.reshape((10000, 1, 28*28))


def binary_decoder(i):
    z = np.zeros((1, 10), dtype=np.float16)
    z[0, i] = 1.0
    return z


y_train = np.array([binary_decoder(i) for i in y_train], dtype=np.float16)
y_test = np.array([binary_decoder(i) for i in y_test], dtype=np.float16)

print(y_train.shape, x_train.shape)
```

```python

output_ = model.fit(x_train, y_train, epochs=5)
```

```python
_output = model.predict_on_batch(x_test)


result = [np.argmax(e) for e in _output]

check = True
for i, e, o in zip(x_test, y_test, result):
    e_=np.argmax(e[0])
    check &= e_ == o
    print("expected : ", e_, " result : ", o, " check: ", e_ == o)
print(check)
```

```python
loss, acc = model.evaluate(x_test, y_test)
print(loss, acc)
```
